```python
# Phân tích tình trạng chứng khoán của Việt Nam thời gian gần dựa trên các headlines từ vnexpress.net
# Import libraries
from urllib.request import urlopen
from bs4 import BeautifulSoup
import os
```

```python
url = "https://vnexpress.net/kinh-doanh/chung-khoan"
html = urlopen(url)
# Read the contents of the file into 'soup'
soup = BeautifulSoup(html, 'lxml')

```

```python
# Find 'article' in the Soup and load it into 'html_table'
html = soup.find_all("article")
html
```

```python
links = []
for item in soup.findAll('h2'):
    link = item.find('a')
    if link:
        links.append(link.text)

print(links)
```

```python
# NLTK VADER for sentiment analysis
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# New words and values
new_words = {
    'cao nhất':10,
    'tăng': 10,
    'trên': 5,
    'đổ vào': 5,
    'giảm': -10,
    'dứt' : -5,
    'sang ':-5,
    'hạn chế': -10,
    'rơi': -10,
}

# Instantiate the sentiment intensity analyzer with the existing lexicon
vader = SentimentIntensityAnalyzer()
# Update the lexicon
vader.lexicon.update(new_words)
```

```python
import pandas as pd

# Iterate through the headlines and get the polarity scores
scores = [vader.polarity_scores(headline) for headline in links]
# Convert the list of dicts into a DataFrame
scores_df = pd.DataFrame(scores)
```
